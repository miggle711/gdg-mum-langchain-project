## Customising your Model

```python
model = init_chat_model(
    model="gemini-2.5-flash",
    model_provider="google_genai",
    temperature=0.7, # controls randomness of output
    max_tokens=300,
    top_p=0.95 # nucleus sampling
)


response = model.invoke("What's the capital of the Moon?")
print(response.content)
```

## Initialising and invoking an agent

```python
from langchain.agents import create_agent

agent = create_agent(model=model)
```

```python
agent = create_agent(model="claude-sonnet-4-5")
```

```python
agent = create_agent("gpt-5-nano")
```

```python
from langchain.messages import HumanMessage


# agent.invoke expects a list of messages, so we need to wrap our input in a HumanMessage
# HumanMessage is a class that represents a message from a human user. It has a content attribute that contains the text of the message.
response = agent.invoke(
    {"messages": [HumanMessage(content="What's the capital of the Moon?")]}
)
```

```python
from pprint import pprint

pprint(response)

# AI messages are represented by the AIMessage class, which has a content attribute that contains the text of the message, and a response_metadata attribute that contains additional information about the response, such as the model used, the tokens used, and the latency of the response.
```

```python
print(response['messages'][-1].content)
```

```python
from langchain.messages import AIMessage

response = agent.invoke(
    {"messages": [HumanMessage(content="What's the capital of the Moon?"),
    AIMessage(content="The capital of the Moon is Luna City."),
    HumanMessage(content="Interesting, tell me more about Luna City")]}
)

# response will contain the entire conversation history, including the new user message and the model's response to it. The last message in the list will be the model's response to the latest user message.
# usage_metadata={'input_tokens': 29, 'output_tokens': 296, 'total_tokens': 325, 'input_token_details': {'cache_read': 0}, 'output_token_details': {'reasoning': 286}}
pprint(response)
```

## Streaming Output

```python
for token, metadata in agent.stream(
    {"messages": [HumanMessage(content="Tell me all about Luna City, the capital of the Moon")]},
    stream_mode="messages"
):

    # token is a message chunk with token content
    # metadata contains which node produced the token
    
    if token.content:  # Check if there's actual content
        print(token.content, end="", flush=True)  # Print token
```
