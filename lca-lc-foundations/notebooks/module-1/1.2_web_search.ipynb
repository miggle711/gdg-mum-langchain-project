## Add web search tool

```python
from langchain.tools import tool
from typing import Dict, Any
from tavily import TavilyClient

# we need Tavily client to access the search tool, which is a wrapper around the tavily API. 
# This allows us to use the search tool in our agent, and the agent will learn to call this tool when it needs to search the web for information.
tavily_client = TavilyClient()

@tool
def web_search(query: str) -> Dict[str, Any]:

    """Search the web for information"""

    return tavily_client.search(query)

web_search.invoke("Who is the current mayor of San Francisco?")
```

```python
from langchain.messages import HumanMessage


agent = create_agent(
    model="gemini-2.5-flash",
    tools=[web_search]
)

question = HumanMessage(content="Who is the current mayor of San Francisco?")

response = agent.invoke(
    {"messages": [question]}
)
```

```python
from pprint import pprint

pprint(response['messages'])
```
